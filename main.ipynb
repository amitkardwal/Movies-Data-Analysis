```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
##Load and Explore the Dataset

df=pd.read_csv("IMDB-Movie-Data.csv")
```

```python
print("First 5 Rows :")
df.head()
```

```python
print("Last 5 Rows :")
df.tail()
```

```python
print(df.shape)
print("Total Rows :",df.shape[0])
print("Total cols: ",df.shape[1])
```

```python
print(df.columns)
```

```python
## Clean the Data

# drop null rows..
df=df.dropna(subset=['Rank', 'Title', 'Genre', 'Description', 'Director', 'Actors', 'Year',
       'Runtime (Minutes)', 'Rating', 'Votes', 'Revenue (Millions)',
       'Metascore'])
df.shape[0]
df.reset_index()
```

```python
## check is all rows are not empty
df.info()
df.describe()
```

```python
##.....................Exploratory Data Analysis (EDA)................................##
```

```python
##Top 10 Movies by Rating
print("Top 10 Rated Movies:")
movie_top=df.sort_values(by="Rating",ascending=False)[['Title',"Rating"]].reset_index().head(5)
print(movie_top)
```

```python
 ##Top 10 Highest Revenue Movies
print("Top 10 Highest Revenue Movies : ")
Movie_Revenue=df.sort_values(by="Revenue (Millions)",ascending=False)[["Title","Revenue (Millions)"]].head(5)
print(Movie_Revenue)
```

```python
## Average Values for Numeric Columns

print("Average stas.... : ")
avg_stas=df[['Runtime (Minutes)', 'Rating', 'Votes', 'Revenue (Millions)']].mean()
print(avg_stas)
```

```python
## latest movie according to data set...
new_movie=df.sort_values(by="Year", ascending=False)[["Title","Year"]].head(5)
print("Top 5 latest Movie : \n",new_movie)

```

```python
## Total Number of Unique Directors
print("Total Unique Director : ",df["Director"].nunique())
```

```python
## Director Who Released the Most Movies
max_dir=df["Director"].value_counts().max()
top_director = df["Director"].value_counts().idxmax()
print(f"Director with Most Movies: {top_director}({max_dir})")
```

```python
##Top 5 Directors by Movie Count
top_dir=df["Director"].value_counts().head(5)
print(top_dir)
```

```python
print(df.columns)
```

```python
##Top 10 Longest Movies by Runtime
longest_movies =df.sort_values(by="Runtime (Minutes)",ascending=False)[["Title","Runtime (Minutes)"]].head(10)
print("Longest Movvie Top 10: \n",longest_movies)
```

```python
## Top 10 Highest Vote count Movie..
print("Top 10 Highest Vote count Movie : ")
max_vote=df.sort_values(by="Votes",ascending=False)[["Title","Votes"]].head(10)
print(max_vote)
```

```python
df.columns

##Most Popular Movie Each Year (by Votes)
popular_movie=df.loc[df.groupby("Year")["Votes"].idxmax()][["Year","Title","Votes"]]
print("Most Popular Movie Each Year (by Votes) : ")
print(popular_movie)
```

```python
##Most Popular Movie Each Year (by Revenue)
popular_movie_Rev=df.loc[df.groupby("Year")["Revenue (Millions)"].idxmax()][["Year","Title","Revenue (Millions)"]]
print("Most Popular Movie Each Year (by Revenue) : ")
print(popular_movie_Rev)
```

```python
## Average Rating per Year
avg_rat_per_year=df.groupby("Year")["Rating"].mean()
print("Average Rating per Year : ")
print(avg_rat_per_year)
```

```python
##  Total Movies Released Per Year
movies_per_year=df.groupby("Year")["Title"].nunique()
print(movies_per_year)
```

```python
##Most Successful Directors (by Average Revenue)
director_rev=df.groupby("Director")["Revenue (Millions)"].mean().sort_values(ascending=False).head(10)
print(director_rev)
```

```python
## Year-wise Total Revenue
total_rev_year =df.groupby("Year")["Revenue (Millions)"].sum()
print(" Total Revenue per Year:")
print(total_rev_year)
```

```python
##"......###&...................Graph Visulization.................$$$$$$$$###########"
```

```python
## Year-wise Total Revenue
total_rev_year.plot(kind="bar",color="red",edgecolor="black")
plt.title("Total Revenue per Year (In Millions) ")
plt.xlabel("Year")
plt.ylabel("Rupees (In Millions)")
plt.tight_layout()
plt.grid(True)

```

```python
 ##  Movies Released per Year 
movies_per_year = df['Year'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
plt.bar(movies_per_year.index, movies_per_year.values, color='skyblue')
plt.title(' Movies Released per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.grid(True)
```

```python
##Average Rating per Year (Line Chart)
avg_rating_year=df.groupby("Year")["Rating"].mean()
plt.bar(avg_rating_year.index,avg_rating_year.values,color="green")
plt.title("Average Rating per Year ")
plt.xlabel("Year")
plt.ylabel("Rating")
plt.tight_layout()
```

```python
## Top 10 Genres by Average Revenue
avg_rev_genre=df.groupby("Genre")["Revenue (Millions)"].mean().sort_values(ascending=False).head(10)

avg_rev_genre.plot(kind="bar",color="red",edgecolor="black")
plt.xlabel("Genre")
plt.ylabel("Revenue (Millions)")
plt.title("Top 10 Genres by Average Revenue")
plt.tight_layout()
plt.grid(True)
```

```python
##Votes vs Rating 
plt.figure(figsize=(9, 5))
plt.scatter(df['Votes'], df['Rating'], alpha=0.5, color='purple')
plt.title('Votes vs  Rating')
plt.xlabel('Votes')
plt.ylabel('Rating')
plt.grid(True)

```

```python
##Top 10 Directors by revenue
plt.figure(figsize=(10, 5))
plt.bar(director_rev.index ,director_rev.values,color="skyblue")
plt.title("Top 10 Directors")
plt.xlabel("director")
plt.ylabel("Revenue")
plt.xticks(rotation=45)

plt.grid(True)

```

```python
##Line Plot: Year vs Avg Revenue
avg_rev_year = df.groupby('Year')['Revenue (Millions)'].mean()
plt.plot(avg_rev_year.index,avg_rev_year.values,marker="*")
plt.title('Avg Revenue Per Year')
plt.xlabel('Year')
plt.ylabel('Revenue (Millions)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
##Pie Chart: Movie Share by Genre
top_genres=df["Genre"].value_counts().head(5)
plt.pie(top_genres,labels=top_genres.index,autopct='%1.1f%%',shadow=True)
plt.title(' Top 5 Genres by Number of Movies')
plt.tight_layout()
```
